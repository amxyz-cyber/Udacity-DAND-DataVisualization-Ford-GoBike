## Investigation Overview

> In this investigation, I'm most interested in figuring out what features are best for predicting when a bike ride will most likely take place and what affects the duration of a bike ride. In particular, I'm going to take the following influences into account: start time, customers as well as the stations.

## Dataset Overview

> The data consists of more than 2.5 million trips taken by shared bikes, including start hour, start weekday, start month, station ids, customer types and other qualities related to a bike sharing system in San Francisco Bay area. However, due to data wrangling more than half million data points were removed.

```python
# To datetime
def my_to_datetime(d,col):
    d[col] = pd.to_datetime(d[col])
```

```python
# to category
def to_category(d,col):
    d[col] = d[col].astype('category')
```

```python
# creates log scaled bins
def get_log_bins(x,col,log_binsize):
    bins = 10 ** np.arange(2.4, np.log10(x[col].max())+log_binsize, log_binsize)
    return bins
```

```python
# creates a heat plot
def draw_heatplot(x,xval,yval,xt,yt,t):
    plt.hist2d(data =x, x = xval, y = yval, cmap='viridis_r',norm=mpl.colors.LogNorm())
    plt.colorbar()
    plt.xlabel(xt)
    plt.ylabel(yt)
    plt.title(t);
```

```python
# creates log scaled histogram
def get_log_histogram(bins,x_size,y_size,x,col,xt,yt,t,ax,ticks_1,ticks_2):
    plt.figure(figsize=[x_size, y_size])
    plt.hist(data = x, x = col, bins = bins)
    plt.xlabel(xt)
    plt.ylabel(yt)
    plt.axis(ax)
    plt.title(t)
    plt.xscale('log')
    if len(ticks_1)>0 and len(ticks_2)>0:
        plt.xticks(ticks_1, ticks_2)
    plt.show()
```

```python
# plots a bar chart
def draw_bar_chart(p,xval,x,l,c):
    sb.set_color_codes(p)
    sb.countplot(x=xval,data=x,label=l,color=c)
```

```python
# creates standard bins
def get_standard_bins(min_value,x,col,binsize):
    bins = np.arange(min_value, x[col].max()+binsize, binsize)
    return bins
```

```python
# creates standard histogram
def get_standard_histogram(bins,x_size,y_size,x,col,xt,yt,t,ax):
    plt.figure(figsize=[x_size, y_size])
    plt.hist(data = x, x = col, bins = bins)
    plt.xlabel(xt)
    plt.ylabel(yt)
    if len(ax) > 0:
        plt.axis(ax)
    plt.title(t)
    plt.show()
```

```python
# creates a violin plot
def draw_violin_plot(x,xval,yval,c,xt,yt,t):
    sb.violinplot(data=x, x=xval, y=yval, color=c,inner='quartile')
    plt.xticks(rotation=20)
    plt.xlabel(xt)
    plt.ylabel(yt)
    plt.title(t);
```

```python
# plot regression 
def draw_regression(xval,yval,x,cat,p,a,l,xt,yt,t):
    sb.lmplot( x=xval, y=yval, data=x, fit_reg=False, hue=cat, legend=False, palette=p, scatter_kws={"alpha":a,"s":15} )
    plt.legend(loc=l, markerscale=2)
    plt.xlabel(xt)
    plt.ylabel(yt)
    plt.title(t, loc='left')
    plt.show()
```

## Distribution of Duration in Seconds

> The duration in seconds ranges from 6 seconds up to 86,000 seconds just for a single trip. As this dataset is logarithmic in nature, so is the duration in seconds with a highly skewed data to the right.

```python
# log-scaled plot of the duration
bin_size=0.025
ax1=[250, 15000, 0, 110000]
bins=get_log_bins(df,ds,bin_size)
ticks_1=[500, 1e3, 2e3, 5e3, 1e4]
ticks_2=[500, '1k', '2k', '5k', '10k']
get_log_histogram(bins,20,10,df,ds,dst,nt,t11,ax1,ticks_1,ticks_2)
```

## Distribution of the Start Hour

> The distribution of the hour at which users of the bike-sharing company started a trip, takes on the shape of a curve. It shows that most trips were taken in the morning or between late afternoon and early evening hours.

```python
# plots the distribution of the start hour
sb.countplot(data=df,palette="vlag", x=sh)
plt.ylabel(nt)
plt.xlabel(hour)
plt.title(t6);
```

## Duration in Seconds vs Hour
> Since both variables are logarithmic in nature, this is reflected through the heatmap. Trips that start in the morning or the late afternoon tend to increase the duration of a trip.

```python
# plots duration vs hour
draw_heatplot(df,sh,ds,hour,dst,t14)
```

## Duration by Hours and Customers
> We've already seen that if a trip starts at a certain hour, it's likely that the duration of a trip increases. When taking the users into account, the diagram shows that there is no linear relationship between these variables. However, customers tend to have longer trips than subscribers.

```python
# plots a regression model consisting of duration in seconds, hours and customer types
draw_regression(sh,ds,df,ut,'deep',0.7,'upper left',hour,dst,t1)
```

## Distribution of the Start Station ID

## Duration vs. Difference in Stations
>  Plotting the duration in seconds against the difference between start and end station ids, the relationship between these two variables becomes clear only on a logarithmic scale. It shows that most trips are short trips or pick up and return station are nearby as the difference in start and end station id for most trips is small. Consequently, the difference in station ids does have a significant influence on the duration, as a small difference in station ids leads to more trips that take longer than on average.

> As pointed out previously, that a small difference in station ids increases the duration in seconds. This is still valid, when looking at start station ids. ## Duration by Weekday and Customers
> If looking at the duration in seconds by taking the weekdays and the customers into account, it becomes clear that the mean of the duration in seconds highly depends on the day of the week as well as the user type. Customers who started their trips on weekends had longer trips on average compared to subscribers who rented their bikes for a much shorter period of time.

```python
# plotting multiple charts at once
fig, axes = plt.subplots(1,2, figsize=(20, 8))

# multivariate plot of duration in seconds by weekdays and user types
sb.set_theme(style="whitegrid")
sb.pointplot(ax=axes[0],data = df, x = swn, y = ds, hue = ut,dodge = 0.3, linestyles = "",legend=False,palette = 'deep')
axes[0].set(ylabel=dst,xlabel=wd)
axes[0].legend(loc='upper left', markerscale=1)
axes[0].tick_params(axis='x', labelrotation=20 )
axes[0].set_title(t22)


# multivariate plot of average duration by weekdays and user types
val='duration_sec_avg'
means = df.groupby([swn,ut]).mean()[ds]
means = means.reset_index(name = val)
means = means.pivot(index = swn, columns = ut, values = val)
sb.heatmap(ax=axes[1],data=means, annot = True, fmt = '.0f', cbar_kws = {'label' : 'mean(duration in seconds)'})
axes[1].set_xlabel(cat_customer)
axes[1].set_ylabel(wd)
axes[1].set_title(t21)


fig.suptitle(t30,fontsize=16)
fig.show()
```
